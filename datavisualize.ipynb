```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,10])
y=np.array([20,100])
print(plt.plot(x,y))
print(plt.show())
```

```python
a=np.array([0,5])
b=np.array([0,25])
print(plt.plot(a,b,marker='p'))
print(plt.show())
```

```python
x=np.array([2,2,2])
y=np.array([1,2,3])
a=np.array([3,2,3])
b=np.array([1,2,3])
c=np.array([1.25,1.25,1.25,0.75])
d=np.array([1,3,1,1])
e=np.array([0.75,1.75])
f=np.array([3,3])
print(plt.plot(x,y,a,b,c,d,e,f))
print(plt.show())
```
